```python
# =========================================
# Transformer-Based AP Selection (Uplink Only) with Rate CDF
# =========================================

import numpy as np
from scipy.io import loadmat
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# -------------------------------
# Step 1: Load Uplink Channel Gain Matrix (BETAA)
# -------------------------------
mat_data = loadmat('BETAA.mat')
beta_matrix = mat_data['BETAA']  # shape: (num_aps, num_users)
beta_matrix = beta_matrix.T      # reshape to (num_users, num_aps)

num_users, num_aps = beta_matrix.shape
top_k = 3  # Each user will be served by their own best top_k APs. This has been selected to 3 assuming the user to be within the triangle boundary; the algorithm we use in graphics.

# Print channel stats for reference
print(f"Channel matrix shape: {beta_matrix.shape}")
print(f"Min channel gain: {np.min(beta_matrix):.4f}, Max channel gain: {np.max(beta_matrix):.4f}")
print(f"Mean channel gain: {np.mean(beta_matrix):.4f}")

# -------------------------------
# Step 2: Normalize Input Features
# -------------------------------
scaler = StandardScaler()
x_normalized = scaler.fit_transform(beta_matrix)

# -------------------------------
# Step 3: Generate Per-User Ground Truth
# -------------------------------
def generate_ground_truth(beta, top_k=3):
    """
    For each user, set 1 for the indices of top-k APs with the highest gain.
    This is the true serving set of APs for each user.
    """
    ground_truth = np.zeros_like(beta)
    for i, row in enumerate(beta):
        top_indices = np.argsort(row)[-top_k:]
        ground_truth[i, top_indices] = 1
    return ground_truth

y_ground_truth = generate_ground_truth(beta_matrix, top_k)

# -------------------------------
# Step 4: Prepare Train/Test Sets
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    x_normalized, y_ground_truth, test_size=0.2, random_state=42
)

X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64)

# -------------------------------
# Step 5: Transformer Model Definition
# -------------------------------
class TransformerModel(nn.Module):
    def __init__(self, input_dim, d_model=64, nhead=4, num_layers=2, dropout=0.1):
        super(TransformerModel, self).__init__()
        self.input_proj = nn.Linear(input_dim, d_model)
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model, nhead=nhead, dropout=dropout, batch_first=True
        )
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.output_proj = nn.Linear(d_model, input_dim)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.input_proj(x).unsqueeze(1)     # (batch_size, seq_len=1, d_model)
        x = self.transformer(x)                 # (batch_size, 1, d_model)
        x = self.output_proj(x.squeeze(1))      # (batch_size, input_dim)
        return self.sigmoid(x)

# -------------------------------
# Step 6: Training Setup
# -------------------------------
model = TransformerModel(input_dim=num_aps)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# -------------------------------
# Step 7: Train the Model
# -------------------------------
train_losses, test_losses = [], []
num_epochs = 50

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    train_losses.append(running_loss / len(train_loader))

    # Evaluate on test set
    model.eval()
    test_loss = 0.0
    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            test_loss += loss.item()
    test_losses.append(test_loss / len(test_loader))

    print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_losses[-1]:.4f}, Test Loss: {test_losses[-1]:.4f}")

# -------------------------------
# Step 8: Final Prediction - Top-k Per User
# -------------------------------
model.eval()
with torch.no_grad():
    all_preds = model(torch.tensor(x_normalized, dtype=torch.float32)).numpy()

binary_preds = np.zeros_like(all_preds)
for i in range(all_preds.shape[0]):
    top_indices = np.argsort(all_preds[i])[-top_k:]
    binary_preds[i, top_indices] = 1

# -------------------------------
# Step 9: Bit-Wise Accuracy Calculation
# -------------------------------
bitwise_accuracy = (binary_preds == y_ground_truth).sum() / y_ground_truth.size * 100
print(f"\nFinal Bit-Wise Accuracy (User-wise Top-{top_k} AP Selection): {bitwise_accuracy:.2f}%")

# =========================================
# Achievable Rate Calculation (Uplink)
# =========================================
def calculate_rates(selection_matrix, beta_matrix, top_k, p=0.1, noise_power=1e-9):
    """
    Calculate achievable rates for all users based on AP selection

    Args:
        selection_matrix: Binary matrix (num_users x num_aps) indicating AP selection
        beta_matrix: Channel gain matrix (num_users x num_aps)
        top_k: Number of APs serving each user
        p: Transmission power per user (Watts)
        noise_power: Noise power (Watts)

    Returns:
        rates: Achievable rates for all users (numpy array)
    """
    num_users, num_aps = beta_matrix.shape
    rates = np.zeros(num_users)

    # For each user k
    for k in range(num_users):
        # Get serving AP indices for user k
        serving_aps = np.where(selection_matrix[k] > 0.5)[0]

        # Desired signal power (sum over serving APs)
        signal_power = p * np.sum(beta_matrix[k, serving_aps])

        # Interference calculation (from all other users)
        interference_power = 0.0
        for j in range(num_users):
            if j == k:
                continue  # Skip own signal
            # Interference from user j to user k's serving APs
            interference_power += p * np.sum(beta_matrix[j, serving_aps])

        # SINR and rate calculation
        sinr = signal_power / (interference_power + noise_power)
        rates[k] = np.log2(1 + sinr)

    return rates

# Transmission parameters
p = 0.1       # Transmission power per user (100 mW)
noise_power = 1e-9  # Noise power (W)

# Calculate rates for both selection methods
gt_rates = calculate_rates(y_ground_truth, beta_matrix, top_k, p, noise_power)
pred_rates = calculate_rates(binary_preds, beta_matrix, top_k, p, noise_power)

# Print rate statistics
print("\nRate Statistics (Ground Truth Selection):")
print(f"Min Rate: {np.min(gt_rates):.4f} bps/Hz")
print(f"Median Rate: {np.median(gt_rates):.4f} bps/Hz")
print(f"Max Rate: {np.max(gt_rates):.4f} bps/Hz")
print(f"Mean Rate: {np.mean(gt_rates):.4f} bps/Hz")

print("\nRate Statistics (Predicted Selection):")
print(f"Min Rate: {np.min(pred_rates):.4f} bps/Hz")
print(f"Median Rate: {np.median(pred_rates):.4f} bps/Hz")
print(f"Max Rate: {np.max(pred_rates):.4f} bps/Hz")
print(f"Mean Rate: {np.mean(pred_rates):.4f} bps/Hz")

# -------------------------------
# Plot Training Curve
# -------------------------------
plt.figure(figsize=(8, 5))
plt.plot(train_losses, label='Train Loss', linewidth=2)
plt.plot(test_losses, label='Test Loss', linewidth=2)
plt.title("Transformer Training Curve (Uplink Only)")
plt.xlabel("Epoch")
plt.ylabel("Binary Cross-Entropy Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# === Display APs selected for each user ===
print("\n[Predicted AP sets for each user]")
for i in range(num_users):
    top_aps = np.where(binary_preds[i] == 1)[0]
    print(f"User {i+1:2d}: APs -> {top_aps.tolist()}")
```

```python
import seaborn as sns

# =========================================
# Comparison Matrix Visualization
# =========================================
comparison_matrix = np.zeros_like(binary_preds, dtype=float)

for user_idx in range(num_users):
    for ap_idx in range(num_aps):
        pred = binary_preds[user_idx, ap_idx]
        gt = y_ground_truth[user_idx, ap_idx]
        if pred == 1 and gt == 1:
            comparison_matrix[user_idx, ap_idx] = 1.0  # match
        elif pred != gt:
            comparison_matrix[user_idx, ap_idx] = 0.5  # mismatch
        # else it remains 0.0 (both 0)

plt.figure(figsize=(12, 6))
sns.heatmap(
    comparison_matrix,
    cmap=["white", "orange", "green"],
    cbar_kws={'ticks': [0, 0.5, 1], 'label': 'AP Selection'},
    linewidths=0.5, linecolor='gray',
    xticklabels=[f"AP{i}" for i in range(num_aps)],
    yticklabels=[f"User{i}" for i in range(num_users)]
)

plt.title("Comparison of Predicted vs Ground Truth APs (Uplink)")
plt.xlabel("Access Points")
plt.ylabel("Users")
plt.tight_layout()
plt.show()
```

```python
# =========================================
# CDF Plot of Achievable Uplink Rates (Required)
# =========================================
plt.figure(figsize=(10, 6))

# Ground truth CDF
sorted_gt = np.sort(gt_rates)
cdf_gt = np.arange(1, len(sorted_gt)+1) / len(sorted_gt)
plt.plot(sorted_gt, cdf_gt, 'b-', linewidth=2.5, label='Ground Truth AP Selection')

# Predicted CDF
sorted_pred = np.sort(pred_rates)
cdf_pred = np.arange(1, len(sorted_pred)+1) / len(sorted_pred)
plt.plot(sorted_pred, cdf_pred, 'r--', linewidth=2.5, label='Transformer Predicted AP Selection')

# Labels and styling
plt.title('CDF of Achievable Uplink Rates per User', fontsize=15)
plt.xlabel('Achievable Uplink Rate (bps/Hz)', fontsize=13)
plt.ylabel('Cumulative Distribution Function (CDF)', fontsize=13)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(left=0)
plt.tight_layout()
plt.show()
```
