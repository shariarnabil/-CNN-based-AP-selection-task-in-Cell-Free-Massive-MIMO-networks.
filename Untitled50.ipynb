```python
# 1. Import Libraries
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from torch.utils.data import DataLoader, TensorDataset
from scipy.io import loadmat

# Parameters for rate calculations
TRANSMIT_POWER = 1.0  # Watts
NOISE_POWER = 1e-9    # Watts
BANDWIDTH = 20e6      # 20 MHz

# -------------------------------
# 2. Load channel gain matrices
# -------------------------------
# Uplink channel
mat_data_ul = loadmat('BETAA.mat')
beta_matrix_ul = mat_data_ul['BETAA']  # shape: (num_aps, num_users)
signal_strengths_ul = beta_matrix_ul.T  # shape: (num_users, num_aps)
num_users, num_aps = signal_strengths_ul.shape

# Downlink channel
mat_data_dl = loadmat('BETAA_DL.mat')
beta_matrix_dl = mat_data_dl['BETAA']  # shape: (num_aps, num_users)
signal_strengths_dl = beta_matrix_dl.T    # shape: (num_users, num_aps)

# -------------------------------
# 3. Compute AP correlation matrix
# -------------------------------
ap_correlation = np.corrcoef(signal_strengths_ul.T)  # shape: (num_aps, num_aps)

# -------------------------------
# 4. Smart AP selection function
# -------------------------------
def select_aps_smart(user_signal, corr_matrix, num_aps_to_select=3, alpha=0.7):
    num_total_aps = len(user_signal)
    norm_gain = (user_signal - np.min(user_signal)) / (np.max(user_signal) - np.min(user_signal) + 1e-8)
    selected_aps = []
    candidate_aps = list(np.argsort(norm_gain)[::-1])

    while len(selected_aps) < num_aps_to_select and candidate_aps:
        scores = []
        for ap in candidate_aps:
            avg_corr = np.mean([abs(corr_matrix[ap, s]) for s in selected_aps]) if selected_aps else 0
            score = alpha * norm_gain[ap] - (1 - alpha) * avg_corr
            scores.append(score)
        best_idx = np.argmax(scores)
        selected_ap = candidate_aps[best_idx]
        selected_aps.append(selected_ap)
        candidate_aps.remove(selected_ap)
    return selected_aps

# -------------------------------
# 5. Select APs for each user
# -------------------------------
X_selected = []
Y_target = []

for user in range(num_users):
    user_signal = signal_strengths_ul[user]
    selected_indices = select_aps_smart(user_signal, ap_correlation, num_aps_to_select=3, alpha=0.7)
    X_selected.append(user_signal[selected_indices])
    Y_target.append(user_signal)

X_selected = np.array(X_selected)
Y_target = np.array(Y_target)

# -------------------------------
# 6. Normalize Data (Z-score)
# -------------------------------
scaler_X = StandardScaler()
scaler_Y = StandardScaler()

X_selected_scaled = scaler_X.fit_transform(X_selected)
Y_target_scaled = scaler_Y.fit_transform(Y_target)

X_tensor = torch.tensor(X_selected_scaled, dtype=torch.float32)
Y_tensor = torch.tensor(Y_target_scaled, dtype=torch.float32)
dataset = TensorDataset(X_tensor, Y_tensor)
train_loader = DataLoader(dataset, batch_size=32, shuffle=True)

# -------------------------------
# 7. Define Regressor Network
# -------------------------------
class APRegressorMLP(nn.Module):
    def __init__(self, input_dim=3, output_dim=num_aps):
        super(APRegressorMLP, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(input_dim, 16),
            nn.ReLU(),
            nn.Linear(16, 64),
            nn.ReLU(),
            nn.Linear(64, output_dim)
        )

    def forward(self, x):
        return self.fc(x)

# -------------------------------
# 8. Train the Model
# -------------------------------
model = APRegressorMLP()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.MSELoss()

loss_history = []

for epoch in range(100):
    epoch_loss = 0
    for xb, yb in train_loader:
        preds = model(xb)
        loss = loss_fn(preds, yb)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    avg_loss = epoch_loss / len(train_loader)
    loss_history.append(avg_loss)
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {avg_loss:.6f}")

# -------------------------------
# 9. Evaluation - Top-5 Accuracy
# -------------------------------
correct_count = 0

with torch.no_grad():
    predicted_scaled = model(X_tensor).numpy()
    predicted_ap_gains = scaler_Y.inverse_transform(predicted_scaled)

for i in range(num_users):
    true_best_ap = np.argmax(Y_target[i])
    pred_top5_aps = np.argsort(predicted_ap_gains[i])[::-1][:5]
    if true_best_ap in pred_top5_aps:
        correct_count += 1

coverage_percent = (correct_count / num_users) * 100
print(f"\nTop-5 APs from prediction cover the actual best AP in {coverage_percent:.2f}% of users")

# =============================================================================
# NEW METRICS CALCULATIONS
# =============================================================================

# -------------------------------
# 10. Throughput and Rate Calculations
# -------------------------------
def calculate_sinr(channel_gains, serving_ap_indices):
    """Calculate SINR for a user given serving AP indices"""
    total_gain = np.sum(channel_gains[serving_ap_indices])
    return (TRANSMIT_POWER * total_gain) / NOISE_POWER

def calculate_rate(sinr):
    """Calculate data rate using Shannon formula"""
    return BANDWIDTH * np.log2(1 + sinr)

# Initialize storage arrays
downlink_rates = np.zeros(num_users)
uplink_rates = np.zeros(num_users)
downlink_throughputs = np.zeros(num_users)
uplink_throughputs = np.zeros(num_users)
effective_aps = np.zeros(num_users)

# For each user, calculate metrics
for user_idx in range(num_users):
    # Get serving APs for this user (selected via smart selection)
    serving_aps = select_aps_smart(signal_strengths_ul[user_idx], ap_correlation)
    effective_aps[user_idx] = len(serving_aps)

    # Uplink calculations (using UL channel gains)
    sinr_ul = calculate_sinr(signal_strengths_ul[user_idx], serving_aps)
    uplink_rates[user_idx] = calculate_rate(sinr_ul)
    uplink_throughputs[user_idx] = uplink_rates[user_idx]  # Throughput = rate in this model

    # Downlink calculations (using DL channel gains)
    sinr_dl = calculate_sinr(signal_strengths_dl[user_idx], serving_aps)
    downlink_rates[user_idx] = calculate_rate(sinr_dl)
    downlink_throughputs[user_idx] = downlink_rates[user_idx]  # Throughput = rate in this model

# =============================================================================
# IMPROVED PLOTTING SECTION
# =============================================================================

# -------------------------------
# 11. Enhanced Plotting Functions
# -------------------------------
def plot_rate_cdf(data, title, xlabel):
    """Plot CDF for rate/throughput metrics with log scale"""
    plt.figure(figsize=(10, 6))
    sorted_data = np.sort(data)
    cdf = np.arange(1, len(sorted_data)+1) / len(sorted_data)

    plt.plot(sorted_data, cdf, linewidth=2.5)
    plt.xscale('log')
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel('CDF', fontsize=12)
    plt.title(title, fontsize=14)
    plt.grid(True, which="both", ls="--")
    plt.minorticks_on()
    plt.tight_layout()
    plt.show()

def plot_cdf_aps(data, title, xlabel):
    """Plot CDF for AP count with integer ticks"""
    plt.figure(figsize=(10, 6))
    sorted_data = np.sort(data)
    cdf = np.arange(1, len(sorted_data)+1) / len(sorted_data)

    plt.plot(sorted_data, cdf, 'r-', linewidth=2.5)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel("CDF", fontsize=12)
    plt.title(title, fontsize=14)
    plt.grid(True)
    plt.xticks(np.unique(sorted_data))
    plt.tight_layout()
    plt.show()

# -------------------------------
# 12. Generate Rate and Throughput Plots
# -------------------------------

# 1. Downlink Rate CDF
plot_rate_cdf(downlink_rates,
             "CDF of Downlink Rate per User",
             "Downlink Data Rate (bps)")

# 2. Downlink Throughput CDF
plot_rate_cdf(downlink_throughputs,
             "CDF of Downlink Net Throughput per User",
             "Downlink Throughput (bps)")

# 3. Uplink Rate CDF
plot_rate_cdf(uplink_rates,
             "CDF of Uplink Rate per User",
             "Uplink Data Rate (bps)")

# 4. Uplink Throughput CDF
plot_rate_cdf(uplink_throughputs,
             "CDF of Uplink Net Throughput per User",
             "Uplink Throughput (bps)")

# 5. Effective APs CDF
plot_cdf_aps(effective_aps,
            "CDF of Effective APs per User",
            "Number of Serving APs")

# 6. Number of Users Histogram
plt.figure(figsize=(10, 6))
unique, counts = np.unique(effective_aps, return_counts=True)
plt.bar(unique, counts, width=0.8, edgecolor='black')
plt.xlabel('Number of Serving APs', fontsize=12)
plt.ylabel('Number of Users', fontsize=12)
plt.title('Distribution of Serving APs per User', fontsize=14)
plt.grid(True, axis='y')
plt.xticks(unique)
plt.tight_layout()
plt.show()

# -------------------------------
# 13. Print Summary Statistics
# -------------------------------
print("\n" + "="*50)
print("Performance Summary Statistics")
print("="*50)
print(f"Top-5 AP Coverage: {coverage_percent:.2f}%")
print(f"Average Downlink Rate: {np.mean(downlink_rates):.2e} bps")
print(f"Median Downlink Rate: {np.median(downlink_rates):.2e} bps")
print(f"Average Uplink Rate: {np.mean(uplink_rates):.2e} bps")
print(f"Median Uplink Rate: {np.median(uplink_rates):.2e} bps")
print(f"Average Serving APs per User: {np.mean(effective_aps):.2f}")
print(f"Users with 1 AP: {np.sum(effective_aps == 1)}")
print(f"Users with 2 APs: {np.sum(effective_aps == 2)}")
print(f"Users with 3 APs: {np.sum(effective_aps == 3)}")
print("="*50)
```
